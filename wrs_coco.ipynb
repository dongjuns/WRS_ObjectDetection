```python
import os
import json
import cv2

path = os.getcwd()
print(path)
datasetList = os.listdir(os.path.join(path, "dataset"))
print(datasetList)
```

```python
annotationsList = os.listdir(os.path.join(path, "annotations"))

jsonDataList = []
coco_annotations = []
imgs = []
jsonData = {}
datasetList.sort()

annotationsList.sort(key=int)
```

```python
# from 0 ~ 199 train, set count0, datasetList[:200]
# from 200 ~ 250 test, set count 200, datasetList[200:]
count = 0
for dataset in datasetList[:200]:
    img_url = dataset
    fileName = dataset.split(".")[0]
            
    annotations = []
    bboxes = []
    
    img_dict = {"id":int, "width":int, "height":int, "file_name":""}
    
    image_url = os.path.join(path, "dataset", img_url)
    img = cv2.imread(image_url)
    height, width = img.shape[:2]
    img_dict["id"] = count
    img_dict["width"] = width
    img_dict["height"] = height
    img_dict["file_name"] = dataset
    imgs.append(img_dict)
    
    for annotation in annotationsList:
        coco_dict = {"id": count+1000, "image_id": count, "category_id": int, "bbox":"", "iscrowd": 0, "area": int}
        antn = "_".join(annotation.split("_", 2)[:2])
        antnRest = "_".join(annotation.split("_", 2)[2:])
        if len(annotations) == 12:
            break
        elif fileName == antn:
            annotations.append(int(antnRest)+1)
            f = open(os.path.join("./annotations/", annotation), "r")
            bbox = f.read()[:-1]
            bbox_split = bbox.split(",")
            bbox_map = map(int, bbox_split)
            bbox_ = list(bbox_map)
            print("bbox", bbox_)
            
            bbox_[2] = bbox_[2] - bbox_[0]
            bbox_[3] = bbox_[3] - bbox_[1]
            
            print("bbox", bbox_)
            
            coco_dict["category_id"] = int(antnRest)+1
            coco_dict["bbox"] = bbox_
            coco_dict["area"] = bbox_[2] * bbox_[3]
            print("coco dict", coco_dict)
            coco_annotations.append(coco_dict)
            print("coco annotations", coco_annotations)
            bboxes.append(bbox_)
            
            f.close()
        else:
            continue
            
    count += 1

jsonData = {
    "info": {
        "year": 2020,
        "description": "WRS objects"
    },
    "licenses": [
        {
            "id":1,
            "name": "Free",
            "url": "url"
        }
    ],
    "images": imgs,
    "categories": [
        {
            "supercategory": "none",
            "name": "obj1",
            "id": 1
        },
        {
            "supercategory": "none",
            "name": "obj2",
            "id": 2
        },
        {
            "supercategory": "none",
            "name": "obj3",
            "id": 3
        },
        {
            "supercategory": "none",
            "name": "obj4",
            "id": 4
        },
        {
            "supercategory": "none",
            "name": "obj5",
            "id": 5
        },
        {
            "supercategory": "none",
            "name": "obj6",
            "id": 6
        },
        {
            "supercategory": "none",
            "name": "obj7",
            "id": 7
        },
        {
            "supercategory": "none",
            "name": "obj8",
            "id": 8
        },
        {
            "supercategory": "none",
            "name": "obj9",
            "id": 9
        },
        {
            "supercategory": "none",
            "name": "obj10",
            "id": 10
        },
        {
            "supercategory": "none",
            "name": "obj11",
            "id": 11
        },
        {
            "supercategory": "none",
            "name": "12",
            "id": 12
        }
    ],
    "annotations": coco_annotations
}
    
with open("temp.json", "w") as write_file:
    json.dump(jsonData, write_file)
```
